## 3)  A) Make a Decision Tree classifier on the on the iris dataset which is an inbuilt dataset Also find its accuracy and visualize the decision tree by writing a code in Jupyter Notebook.

```python
# Importing libraries
import pandas as pd
from sklearn import datasets
from sklearn import tree
from sklearn import metrics
from sklearn.model_selection import train_test_split
from IPython.display import Image  
import pydotplus
```

```python
# Loading data set
iris = datasets.load_iris() 
```

```python
# Train Test split using Hold Out Method
irisdata_train, irisdata_test, iristarget_train, iristarget_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
# Creating the classifier
DT_model = tree.DecisionTreeClassifier() 

# Training the model
DT_fitted = DT_model.fit(irisdata_train, iristarget_train)
```

```python
# Prediction
DT_predict = DT_fitted.predict((irisdata_test))
```

```python
# Accuracy 
print("Accuracy:",round(metrics.accuracy_score(iristarget_test, DT_predict),2) * 100, "%")
```

```python
# Visuallize the Decision Tree

dot_data = tree.export_graphviz(DT_model, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names)

# Draw graph

graph = pydotplus.graph_from_dot_data(dot_data)  

# Show graph
Image(graph.create_png())
```
