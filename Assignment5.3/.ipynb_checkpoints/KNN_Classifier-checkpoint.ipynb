## 3) B) Make a K Nearest Neighbor classifier on the same dataset as above (iris dataset). Evaluate the algorithm using confusion matrix and compare the error rate with the K value and give the best range of K with lowest error rate by writing the code in Jupyter Notebook

```python
# Importing Libraries
import pandas as pd
from sklearn import datasets 
from sklearn.neighbors import KNeighborsClassifier   
from sklearn import metrics
from sklearn.model_selection import train_test_split 
```

```python
# Loading data set
iris = datasets.load_iris() 
```

```python
# Train Test split using Hold Out Method
irisdata_train, irisdata_test, iristarget_train, iristarget_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
# Running the classifier for K range 1 to 25
k_range = range(1, 26)
error_list = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(irisdata_train, iristarget_train)
    knn_pred = knn.predict(irisdata_test)
    confusion = metrics.confusion_matrix(iristarget_test, knn_pred)
    accuracy = metrics.accuracy_score(iristarget_test, knn_pred)
    var1 = "1"
    var1 = int(var1)
    error = var1 - accuracy
    error_list.append(error)
    #error_value = 1-score
    #error.append(error_value)
    print("For K: {}".format(k))
    print("Confusion Matrix-----------------------------------")
    print(confusion)
    print("---------------------------------------------------")
    print("Error Rate-------------------------------------------")
    print(error)
    print("---------------------------------------------------")
    #print(error)

print("Error Rate List for K = 1 to 25")

for i in range(len(error_list)): 
    print ("K= ", i, end = " ") 
    print ("->",error_list[i]) 
```
